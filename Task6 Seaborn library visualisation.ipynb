```python
# Example
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
x=[1,15,9,6,11,3,7,4]
y=[20,8,4,12,19,2,6,1]
df=pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```

```python
# Example
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
x=[1,15,9,6,11,3,7,4]
y=[20,8,4,12,19,2,6,1]
df=pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df,ls=':')
plt.show()
```

```python
# Use a dashed line:
# Example
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
x=[1,15,9,6,11,3,7,4]
y=[20,8,4,12,19,2,6,1]
df=pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df,ls='--')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.array([1,3,7,9,4,5,6,2])
y=np.array([7,6,4,8,9,12,14,20])
df=pd.DataFrame(x,y)
sns.heatmap(df)
plt.show()
```

```python
# Pair plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('iris')
graph=sns.FacetGrid(df)
graph.map(plt.scatter,'sepal_length','petal_length')
plt.show()
```

```python
# Styling plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('iris')
sns.set_style('dark')
plt.figure(figsize=(12,3))
sns.boxplot(x='species',y='petal_length',data=df)
sns.despine()
plt.grid()
plt.show()
```
